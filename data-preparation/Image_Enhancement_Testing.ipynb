# Image Enhancement Testing

## Imports

```python
# Add Utils to path
import sys
sys.path.append('../Utils')  # Adds higher directory to python modules path.

# Image Enhancement Functions
from image_enhancement_functions import histogram_equalization, clahe, color_balance_adjustment

# Other Imports
import cv2
import matplotlib.pyplot as plt
```

## Load Image

```python
# Load an RGB image
image = cv2.imread('../data/ExDark/Bus/2015_02353.jpg')
```

## Histogram equalization

```python
# Apply
equalized_rgb_image = histogram_equalization(image)

# Display the original and equalized images
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Enhanced')
plt.axis('off')
plt.show()
```

## CLAHE

```python
# Apply
enhanced_rgb_image = clahe(image)

# Display the original and equalized images
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(enhanced_rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Enhanced')
plt.axis('off')
plt.show()
```

## Color Balance Adjustment

```python
# Apply 
adjusted_image = color_balance_adjustment(image)

# Display the original and adjusted images
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title('Enhanced')
plt.axis('off')
plt.show()
```
